```
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt
from torch.distributions import Categorical
import soccer_twos

# ===============================
# 1. Actor & Critic Networks
# ===============================
class Actor(nn.Module):
    def __init__(self, obs_dim, action_branches):
        super().__init__()
        self.shared = nn.Sequential(
            nn.Linear(obs_dim, 128),
            nn.ReLU(),
            nn.Linear(128, 128),
            nn.ReLU()
        )
        self.heads = nn.ModuleList([nn.Linear(128, n) for n in action_branches])

    def forward(self, x):
        x = self.shared(x)
        return [torch.softmax(h(x), dim=-1) for h in self.heads]


class Critic(nn.Module):
    def __init__(self, obs_dim, action_dim, num_agents):
        super().__init__()
        input_dim = num_agents * (obs_dim + action_dim)
        self.net = nn.Sequential(
            nn.Linear(input_dim, 256),
            nn.ReLU(),
            nn.Linear(256, 256),
            nn.ReLU(),
            nn.Linear(256, 1)
        )

    def forward(self, obs, acts):
        x = torch.cat([obs.reshape(obs.size(0), -1), acts.reshape(acts.size(0), -1)], dim=1)
        return self.net(x)


# ===============================
# 2. MADDPG Agent
# ===============================
class MADDPG:
    def __init__(self, obs_dim, action_branches, num_agents, device="cpu",
                 gamma=0.95, tau=0.01, lr=1e-3, batch_size=256):
        self.num_agents = num_agents
        self.obs_dim = obs_dim
        self.action_branches = action_branches
        self.action_dim = sum(action_branches)  # 3+3+3 = 9

        self.device = torch.device(device)

        # One actor per agent, one shared critic
        self.actors = [Actor(obs_dim, action_branches).to(self.device) for _ in range(num_agents)]
        self.critics = Critic(obs_dim, self.action_dim, num_agents).to(self.device)

        self.actor_opts = [optim.Adam(a.parameters(), lr=lr) for a in self.actors]
        self.critic_opt = optim.Adam(self.critics.parameters(), lr=lr)

        self.gamma = gamma
        self.tau = tau
        self.batch_size = batch_size

        self.memory = []
        self.actor_losses = [[] for _ in range(num_agents)]
        self.critic_losses = []

    def select_action(self, obs):
        actions, onehots = [], []
        obs = torch.tensor(obs, dtype=torch.float32).to(self.device)

        for i in range(self.num_agents):
            dists = self.actors[i](obs[i])
            branch_actions, branch_onehots = [], []
            for dist in dists:
                dist = Categorical(dist)
                a = dist.sample()
                branch_actions.append(a.item())
                oh = torch.zeros_like(dist.probs)
                oh[a] = 1.0
                branch_onehots.append(oh)
            actions.append(branch_actions)
            onehots.append(torch.cat(branch_onehots))  # shape (9,)
        return np.array(actions), torch.stack(onehots)

    def store(self, transition):
        self.memory.append(transition)

    def learn(self):
        if len(self.memory) < self.batch_size:
            return

        # Sample indices, not objects
        indices = np.random.choice(len(self.memory), self.batch_size, replace=False)
        batch = [self.memory[i] for i in indices]

        obs_batch = torch.tensor(np.array([b[0] for b in batch]), dtype=torch.float32).to(self.device)  # (B,4,336)
        acts_batch = torch.tensor(np.array([b[1] for b in batch]), dtype=torch.float32).to(self.device)  # (B,4,9)
        rews_batch = torch.tensor(np.array([b[2] for b in batch]), dtype=torch.float32).to(self.device)  # (B,4)
        next_obs_batch = torch.tensor(np.array([b[3] for b in batch]), dtype=torch.float32).to(self.device)
        next_acts_batch = torch.tensor(np.array([b[4] for b in batch]), dtype=torch.float32).to(self.device)

        # Critic update
        q_values = self.critics(obs_batch, acts_batch).squeeze()
        with torch.no_grad():
            target_q = self.critics(next_obs_batch, next_acts_batch).squeeze()
            y = rews_batch.mean(dim=1) + self.gamma * target_q
        critic_loss = nn.MSELoss()(q_values, y)

        self.critic_opt.zero_grad()
        critic_loss.backward()
        self.critic_opt.step()
        self.critic_losses.append(critic_loss.item())

        # Actor update per agent
        for i in range(self.num_agents):
            pred_actions = []
            for j in range(self.num_agents):
                dists = self.actors[j](obs_batch[:, j, :])
                branch_acts = []
                for dist in dists:
                    dist = Categorical(dist)
                    a = dist.sample()
                    oh = torch.zeros_like(dist.probs)
                    oh[a] = 1.0
                    branch_acts.append(oh)
                pred_actions.append(torch.cat(branch_acts, dim=-1))
            pred_actions = torch.stack(pred_actions, dim=1)

            actor_loss = -self.critics(obs_batch, pred_actions).mean()

            self.actor_opts[i].zero_grad()
            actor_loss.backward()
            self.actor_opts[i].step()
            self.actor_losses[i].append(actor_loss.item())

        # (Optional) clear buffer every update
        self.memory = []



# ===============================
# 3. Training Loop
# ===============================
env = soccer_twos.make(render=False)
obs_dim = env.observation_space.shape[0]
action_branches = env.action_space.nvec
num_agents = 4

agent = MADDPG(obs_dim, action_branches, num_agents, device="cuda" if torch.cuda.is_available() else "cpu")

episodes = 100
team_rewards = {"blue": [], "orange": []}
blue_win_rate, orange_win_rate = [], []
blue_goals, orange_goals = [], []
cum_blue, cum_orange = 0, 0
blue_wins, orange_wins, total_games = 0, 0, 0

for ep in range(episodes):
    obs = env.reset()
    done, ep_rew = {"__all__": False}, np.zeros(num_agents)

    while not done["__all__"]:
        obs_arr = np.stack([obs[i] for i in range(num_agents)])
        actions, onehots = agent.select_action(obs_arr)

        action_dict = {i: np.array(actions[i]) for i in range(num_agents)}
        next_obs, reward, done, _ = env.step(action_dict)

        next_obs_arr = np.stack([next_obs[i] for i in range(num_agents)])
        _, next_onehots = agent.select_action(next_obs_arr)

        agent.store((obs_arr, onehots.cpu().numpy(), np.array(list(reward.values())),
                     next_obs_arr, next_onehots.cpu().numpy()))

        obs = next_obs
        ep_rew += np.array(list(reward.values()))

    # Team rewards
    blue_total = ep_rew[0] + ep_rew[1]
    orange_total = ep_rew[2] + ep_rew[3]
    team_rewards["blue"].append(blue_total)
    team_rewards["orange"].append(orange_total)

    # Goals
    blue_goal, orange_goal = 0, 0
    if blue_total <= -2: orange_goal = 1
    elif orange_total <= -2: blue_goal = 1
    cum_blue += blue_goal
    cum_orange += orange_goal
    blue_goals.append(cum_blue)
    orange_goals.append(cum_orange)

    # Win rate
    if blue_total > orange_total: blue_wins += 1
    elif orange_total > blue_total: orange_wins += 1
    total_games += 1
    blue_win_rate.append(blue_wins / total_games)
    orange_win_rate.append(orange_wins / total_games)

    agent.learn()
    print(f"Episode {ep+1}: BlueReward={blue_total:.3f}, OrangeReward={orange_total:.3f} "
          f"| CUM BlueGoals={cum_blue}, CUM OrangeGoals={cum_orange}")

env.close()

# ===============================
# 4. Plots
# ===============================
plt.plot(team_rewards["blue"], label="Blue Team Reward")
plt.plot(team_rewards["orange"], label="Orange Team Reward")
plt.xlabel("Episode"); plt.ylabel("Reward")
plt.title("Team Rewards per Episode"); plt.legend(); plt.show()

for i in range(num_agents):
    plt.plot(agent.actor_losses[i], label=f"Actor Loss Agent {i}")
plt.xlabel("Update Step"); plt.ylabel("Actor Loss")
plt.title("Actor Loss per Agent"); plt.legend(); plt.show()

plt.plot(agent.critic_losses, label="Critic Loss")
plt.xlabel("Update Step"); plt.ylabel("Critic Loss")
plt.title("Critic Loss"); plt.legend(); plt.show()

plt.plot(blue_goals, label="Cumulative Blue Goals")
plt.plot(orange_goals, label="Cumulative Orange Goals")
plt.xlabel("Episode"); plt.ylabel("Goals")
plt.title("Cumulative Goals per Team"); plt.legend(); plt.show()

plt.plot(blue_win_rate, label="Blue Team Win Rate")
plt.plot(orange_win_rate, label="Orange Team Win Rate")
plt.xlabel("Episode"); plt.ylabel("Win Rate")
plt.title("Team Win Rates"); plt.legend(); plt.show()
```
