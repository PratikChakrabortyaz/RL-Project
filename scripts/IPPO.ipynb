```
import torch
import torch.nn as nn
import torch.optim as optim
from torch.distributions import Categorical
import numpy as np
import matplotlib.pyplot as plt
import soccer_twos

# ===============================
# 1. Actor-Critic Networks
# ===============================
class Actor(nn.Module):
    def __init__(self, obs_dim, action_branches):
        super().__init__()
        self.shared = nn.Sequential(
            nn.Linear(obs_dim, 128),
            nn.ReLU(),
            nn.Linear(128, 128),
            nn.ReLU()
        )
        self.heads = nn.ModuleList([nn.Linear(128, n) for n in action_branches])

    def forward(self, x):
        x = self.shared(x)
        return [torch.softmax(h(x), dim=-1) for h in self.heads]

class Critic(nn.Module):
    def __init__(self, obs_dim):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(obs_dim, 256),
            nn.ReLU(),
            nn.Linear(256, 256),
            nn.ReLU(),
            nn.Linear(256, 1)
        )

    def forward(self, x):
        return self.net(x)


# ===============================
# 2. IPPO Agent
# ===============================
class IPPO:
    def __init__(self, obs_dim, action_branches, num_agents, device="cpu",
                 gamma=0.99, lr=3e-4, batch_size=512, epochs=5,
                 clip=0.2, ent_coef=0.01, vf_coef=0.5):

        self.num_agents = num_agents
        self.device = torch.device(device)

        self.actors = [Actor(obs_dim, action_branches).to(self.device) for _ in range(num_agents)]
        self.critics = [Critic(obs_dim).to(self.device) for _ in range(num_agents)]
        self.actor_opts = [optim.Adam(a.parameters(), lr=lr) for a in self.actors]
        self.critic_opts = [optim.Adam(c.parameters(), lr=lr) for c in self.critics]

        self.gamma = gamma
        self.batch_size = batch_size
        self.epochs = epochs
        self.clip = clip
        self.ent_coef = ent_coef
        self.vf_coef = vf_coef

        self.memory = [[] for _ in range(num_agents)]
        self.actor_losses = [[] for _ in range(num_agents)]
        self.critic_losses = [[] for _ in range(num_agents)]

    def select_action(self, obs):
        actions, log_probs, values = [], [], []
        obs = torch.tensor(obs, dtype=torch.float32).to(self.device)
        for i in range(self.num_agents):
            dists = self.actors[i](obs[i])
            branch_actions, branch_log_probs = [], []
            for dist in dists:
                dist = Categorical(dist)
                a = dist.sample()
                branch_actions.append(a.item())
                branch_log_probs.append(dist.log_prob(a))
            actions.append(branch_actions)
            log_probs.append(torch.stack(branch_log_probs).sum())
            values.append(self.critics[i](obs[i]))
        return np.array(actions), log_probs, values

    def store(self, agent_idx, transition):
        self.memory[agent_idx].append(transition)

    def learn(self):
        for agent_idx in range(self.num_agents):
            mem = self.memory[agent_idx]
            if len(mem) < self.batch_size:
                continue

            obs = torch.tensor(np.array([m["obs"] for m in mem]), dtype=torch.float32).to(self.device)
            log_probs_old = torch.stack([m["log_prob"] for m in mem]).detach().to(self.device)
            values = torch.cat([m["value"] for m in mem]).squeeze().detach().to(self.device)
            rewards = torch.tensor(np.array([m["reward"] for m in mem]), dtype=torch.float32).to(self.device)
            dones = torch.tensor(np.array([m["done"] for m in mem]), dtype=torch.float32).to(self.device)

            # GAE
            advantages, gae = [], 0
            for t in reversed(range(len(rewards))):
                next_v = 0 if t == len(rewards)-1 else values[t+1]
                delta = rewards[t] + self.gamma * next_v * (1 - dones[t]) - values[t]
                gae = delta + self.gamma * 0.95 * (1 - dones[t]) * gae
                advantages.insert(0, gae)
            advantages = torch.tensor(advantages, dtype=torch.float32).to(self.device)
            returns = advantages + values

            # PPO updates
            for _ in range(self.epochs):
                idx = np.random.permutation(len(mem))
                for start in range(0, len(mem), self.batch_size):
                    batch = idx[start:start+self.batch_size]
                    batch_obs = obs[batch]
                    batch_adv = advantages[batch]
                    batch_ret = returns[batch]
                    batch_log_probs_old = log_probs_old[batch]

                    new_log_probs, entropies = [], []
                    for i in batch:
                        dists = self.actors[agent_idx](torch.tensor(mem[i]["obs"], dtype=torch.float32).to(self.device))
                        lp_per_branch = []
                        for k, dist_probs in enumerate(dists):
                            dist = Categorical(dist_probs)
                            a = torch.tensor(mem[i]["actions"][k]).to(self.device)
                            lp_per_branch.append(dist.log_prob(a))
                            entropies.append(dist.entropy().mean())
                        new_log_probs.append(torch.stack(lp_per_branch).sum())
                    new_log_probs = torch.stack(new_log_probs)

                    ratio = torch.exp(new_log_probs - batch_log_probs_old)
                    surr1 = ratio * batch_adv
                    surr2 = torch.clamp(ratio, 1 - self.clip, 1 + self.clip) * batch_adv
                    actor_loss = -torch.min(surr1, surr2).mean()

                    values_pred = self.critics[agent_idx](batch_obs).squeeze()
                    critic_loss = nn.MSELoss()(values_pred, batch_ret)

                    entropy_loss = torch.stack(entropies).mean()
                    loss = actor_loss + self.vf_coef * critic_loss - self.ent_coef * entropy_loss

                    self.actor_opts[agent_idx].zero_grad()
                    self.critic_opts[agent_idx].zero_grad()
                    loss.backward()
                    self.actor_opts[agent_idx].step()
                    self.critic_opts[agent_idx].step()

                    self.actor_losses[agent_idx].append(actor_loss.item())
                    self.critic_losses[agent_idx].append(critic_loss.item())

            self.memory[agent_idx] = []


# ===============================
# 3. Training Loop (100 Episodes) + Cumulative True Goals (end-of-episode rule)
# ===============================
env = soccer_twos.make(render=False)
obs_dim = env.observation_space.shape[0]
action_branches = env.action_space.nvec
num_agents = 4

agent = IPPO(obs_dim, action_branches, num_agents, device="cuda" if torch.cuda.is_available() else "cpu")

episodes = 100
team_rewards = {"blue": [], "orange": []}

blue_cum_goals, orange_cum_goals = [], []
blue_goals_total, orange_goals_total = 0, 0

blue_win_rate, orange_win_rate = [], []
blue_wins, orange_wins, total_games = 0, 0, 0

GOAL_THRESH = -1.5  # robust threshold for "team conceded a goal" (≈ -2 with small shaping noise)

for ep in range(episodes):
    obs = env.reset()
    done, ep_rew = {"__all__": False}, np.zeros(num_agents)

    while not done["__all__"]:
        obs_arr = np.stack([obs[i] for i in range(num_agents)])
        actions, log_probs, values = agent.select_action(obs_arr)

        action_dict = {i: np.array(actions[i]) for i in range(num_agents)}
        next_obs, reward, done, _ = env.step(action_dict)

        for i in range(num_agents):
            agent.store(i, {
                "obs": obs_arr[i],
                "actions": actions[i],
                "log_prob": log_probs[i],
                "value": values[i],
                "reward": reward[i],
                "done": float(done["__all__"]),
            })

        obs = next_obs
        ep_rew += np.array(list(reward.values()))

    # shaped team totals
    blue_total = ep_rew[0] + ep_rew[1]
    orange_total = ep_rew[2] + ep_rew[3]
    team_rewards["blue"].append(blue_total)
    team_rewards["orange"].append(orange_total)

    # TRUE goal counting by simple rule:
    # team that has approx -2 total conceded a goal → opponent +1 goal
    blue_goals_ep, orange_goals_ep = 0, 0
    if blue_total <= GOAL_THRESH and orange_total >= 0:
        orange_goals_ep = 1
    elif orange_total <= GOAL_THRESH and blue_total >= 0:
        blue_goals_ep = 1
    # else: no goal

    blue_goals_total += blue_goals_ep
    orange_goals_total += orange_goals_ep
    blue_cum_goals.append(blue_goals_total)
    orange_cum_goals.append(orange_goals_total)

    # win rate from shaped totals (optional)
    if blue_total > orange_total:
        blue_wins += 1
    elif orange_total > blue_total:
        orange_wins += 1
    total_games += 1
    blue_win_rate.append(blue_wins / total_games)
    orange_win_rate.append(orange_wins / total_games)

    agent.learn()
    print(f"Episode {ep+1}: BlueReward={blue_total:.3f}, OrangeReward={orange_total:.3f} | "
          f"BlueGoals+={blue_goals_ep}, OrangeGoals+={orange_goals_ep} | "
          f"CUM BlueGoals={blue_goals_total}, CUM OrangeGoals={orange_goals_total}")

env.close()

# ===============================
# 4. Plotting
# ===============================
window = 10
def smooth(x, w):
    return np.convolve(x, np.ones(w)/w, mode="valid") if len(x) >= w else np.array(x, dtype=float)

# (A) Team shaped rewards (smoothed)
if len(team_rewards["blue"]) > 0:
    plt.plot(smooth(team_rewards["blue"], window), label="Blue Team Reward")
    plt.plot(smooth(team_rewards["orange"], window), label="Orange Team Reward")
    plt.xlabel("Episode")
    plt.ylabel("Reward (smoothed)")
    plt.title("Team Rewards")
    plt.legend()
    plt.show()

# (B) Cumulative TRUE goals scored per team (integers)
plt.plot(blue_cum_goals, label="Blue Cumulative Goals")
plt.plot(orange_cum_goals, label="Orange Cumulative Goals")
plt.xlabel("Episodes")
plt.ylabel("Cumulative Goals")
plt.title("Cumulative True Goals Scored per Team (Episode-level rule)")
plt.legend()
plt.show()

# (C) Policy Loss per Agent
for i in range(num_agents):
    plt.plot(agent.actor_losses[i], label=f"Policy Loss Agent {i}")
plt.xlabel("Update Steps")
plt.ylabel("Policy Loss")
plt.title("Policy Loss per Agent")
plt.legend()
plt.show()

# (D) Critic Loss per Agent
for i in range(num_agents):
    plt.plot(agent.critic_losses[i], label=f"Critic Loss Agent {i}")
plt.xlabel("Update Steps")
plt.ylabel("Critic Loss")
plt.title("Critic Loss per Agent")
plt.legend()
plt.show()

# (E) Win rates (from shaped totals; optional)
plt.plot(blue_win_rate, label="Blue Team Win Rate")
plt.plot(orange_win_rate, label="Orange Team Win Rate")
plt.xlabel("Episodes")
plt.ylabel("Win Rate")
plt.title("Team Win Rates")
plt.legend()
plt.show()
```

```
import torch
import torch.nn as nn
import torch.optim as optim
from torch.distributions import Categorical
import numpy as np
import matplotlib.pyplot as plt
import soccer_twos

# ===============================
# 1. Actor-Critic Networks
# ===============================
class Actor(nn.Module):
    def __init__(self, obs_dim, action_branches):
        super().__init__()
        self.shared = nn.Sequential(
            nn.Linear(obs_dim, 128),
            nn.ReLU(),
            nn.Linear(128, 128),
            nn.ReLU()
        )
        self.heads = nn.ModuleList([nn.Linear(128, n) for n in action_branches])

    def forward(self, x):
        x = self.shared(x)
        return [torch.softmax(h(x), dim=-1) for h in self.heads]

class Critic(nn.Module):
    def __init__(self, obs_dim):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(obs_dim, 256),
            nn.ReLU(),
            nn.Linear(256, 256),
            nn.ReLU(),
            nn.Linear(256, 1)
        )

    def forward(self, x):
        return self.net(x)


# ===============================
# 2. IPPO Agent
# ===============================
class IPPO:
    def __init__(self, obs_dim, action_branches, num_agents, device="cpu",
                 gamma=0.99, lr=3e-4, batch_size=512, epochs=5,
                 clip=0.2, ent_coef=0.01, vf_coef=0.5):

        self.num_agents = num_agents
        self.device = torch.device(device)

        # One actor–critic per agent
        self.actors = [Actor(obs_dim, action_branches).to(self.device) for _ in range(num_agents)]
        self.critics = [Critic(obs_dim).to(self.device) for _ in range(num_agents)]
        self.actor_opts = [optim.Adam(a.parameters(), lr=lr) for a in self.actors]
        self.critic_opts = [optim.Adam(c.parameters(), lr=lr) for c in self.critics]

        self.gamma = gamma
        self.batch_size = batch_size
        self.epochs = epochs
        self.clip = clip
        self.ent_coef = ent_coef
        self.vf_coef = vf_coef

        self.memory = [[] for _ in range(num_agents)]  # per-agent memory

        # Logs
        self.actor_losses = [[] for _ in range(num_agents)]
        self.critic_losses = [[] for _ in range(num_agents)]

    def select_action(self, obs):
        actions, log_probs, values = [], [], []
        obs = torch.tensor(obs, dtype=torch.float32).to(self.device)

        for i in range(self.num_agents):
            dists = self.actors[i](obs[i])
            branch_actions, branch_log_probs = [], []
            for dist in dists:
                dist = Categorical(dist)
                a = dist.sample()
                branch_actions.append(a.item())
                branch_log_probs.append(dist.log_prob(a))
            actions.append(branch_actions)
            log_probs.append(torch.stack(branch_log_probs).sum())
            values.append(self.critics[i](obs[i]))

        return np.array(actions), log_probs, values

    def store(self, agent_idx, transition):
        self.memory[agent_idx].append(transition)

    def learn(self):
        for agent_idx in range(self.num_agents):
            mem = self.memory[agent_idx]
            if len(mem) < self.batch_size:
                continue

            obs = torch.tensor(np.array([m["obs"] for m in mem]), dtype=torch.float32).to(self.device)
            actions = [m["actions"] for m in mem]
            log_probs_old = torch.stack([m["log_prob"] for m in mem]).detach().to(self.device)
            values = torch.cat([m["value"] for m in mem]).squeeze().detach().to(self.device)
            rewards = torch.tensor(np.array([m["reward"] for m in mem]), dtype=torch.float32).to(self.device)
            dones = torch.tensor(np.array([m["done"] for m in mem]), dtype=torch.float32).to(self.device)

            # compute returns + advantages
            advantages, gae = [], 0
            for t in reversed(range(len(rewards))):
                delta = rewards[t] + self.gamma * (0 if t == len(rewards)-1 else values[t+1]) * (1-dones[t]) - values[t]
                gae = delta + self.gamma * 0.95 * (1-dones[t]) * gae
                advantages.insert(0, gae)
            advantages = torch.tensor(advantages, dtype=torch.float32).to(self.device)
            returns = advantages + values

            # PPO updates
            for _ in range(self.epochs):
                idx = np.random.permutation(len(mem))
                for start in range(0, len(mem), self.batch_size):
                    batch = idx[start:start+self.batch_size]

                    batch_obs = obs[batch]
                    batch_adv = advantages[batch]
                    batch_ret = returns[batch]
                    batch_log_probs_old = log_probs_old[batch]

                    new_log_probs, entropies = [], []
                    for i in batch:
                        dists = self.actors[agent_idx](torch.tensor(mem[i]["obs"], dtype=torch.float32).to(self.device))
                        lp_per_branch = []
                        for k, dist_probs in enumerate(dists):
                            dist = Categorical(dist_probs)
                            a = torch.tensor(mem[i]["actions"][k]).to(self.device)
                            lp_per_branch.append(dist.log_prob(a))
                            entropies.append(dist.entropy().mean())
                        new_log_probs.append(torch.stack(lp_per_branch).sum())
                    new_log_probs = torch.stack(new_log_probs)

                    # PPO loss
                    ratio = torch.exp(new_log_probs - batch_log_probs_old)
                    surr1 = ratio * batch_adv
                    surr2 = torch.clamp(ratio, 1-self.clip, 1+self.clip) * batch_adv
                    actor_loss = -torch.min(surr1, surr2).mean()

                    values_pred = self.critics[agent_idx](batch_obs).squeeze()
                    critic_loss = nn.MSELoss()(values_pred, batch_ret)

                    entropy_loss = torch.stack(entropies).mean()
                    loss = actor_loss + self.vf_coef*critic_loss - self.ent_coef*entropy_loss

                    self.actor_opts[agent_idx].zero_grad()
                    self.critic_opts[agent_idx].zero_grad()
                    loss.backward()
                    self.actor_opts[agent_idx].step()
                    self.critic_opts[agent_idx].step()

                    # log losses
                    self.actor_losses[agent_idx].append(actor_loss.item())
                    self.critic_losses[agent_idx].append(critic_loss.item())

            self.memory[agent_idx] = []


# ===============================
# 3. Training Loop (IPPO)
# ===============================
env = soccer_twos.make(render=False)
obs_dim = env.observation_space.shape[0]
action_branches = env.action_space.nvec
num_agents = 4

agent = IPPO(obs_dim, action_branches, num_agents, device="cuda" if torch.cuda.is_available() else "cpu")

episodes = 10  # increase for longer runs
agent_rewards = [[] for _ in range(num_agents)]
blue_win_rate, orange_win_rate = [], []
blue_wins, orange_wins, total_games = 0, 0, 0

for ep in range(episodes):
    obs = env.reset()
    done, ep_rew = {"__all__": False}, np.zeros(num_agents)

    while not done["__all__"]:
        obs_arr = np.stack([obs[i] for i in range(num_agents)])
        actions, log_probs, values = agent.select_action(obs_arr)

        action_dict = {i: np.array(actions[i]) for i in range(num_agents)}
        next_obs, reward, done, _ = env.step(action_dict)

        for i in range(num_agents):
            agent.store(i, {
                "obs": obs_arr[i],
                "actions": actions[i],
                "log_prob": log_probs[i],
                "value": values[i],
                "reward": reward[i],
                "done": float(done["__all__"]),
            })

        obs = next_obs
        ep_rew += np.array(list(reward.values()))

    # log rewards
    for i in range(num_agents):
        agent_rewards[i].append(ep_rew[i])

    # win rate tracking
    if sum(ep_rew[:2]) > sum(ep_rew[2:]):
        blue_wins += 1
    elif sum(ep_rew[:2]) < sum(ep_rew[2:]):
        orange_wins += 1
    total_games += 1
    blue_win_rate.append(blue_wins / total_games)
    orange_win_rate.append(orange_wins / total_games)

    agent.learn()
    print(f"Episode {ep+1}: {ep_rew}")

env.close()

# ===============================
# 4. Plotting
# ===============================

# Rewards
for i in range(num_agents):
    plt.plot(agent_rewards[i], label=f"Agent {i}")
plt.xlabel("Episode")
plt.ylabel("Reward")
plt.title("Per-Agent Rewards")
plt.legend()
plt.show()

# Losses (average per episode)
for i in range(num_agents):
    plt.plot(agent.actor_losses[i], label=f"Policy Loss Agent {i}")
plt.xlabel("Update Steps")
plt.ylabel("Policy Loss")
plt.title("Policy Loss per Agent")
plt.legend()
plt.show()

for i in range(num_agents):
    plt.plot(agent.critic_losses[i], label=f"Critic Loss Agent {i}")
plt.xlabel("Update Steps")
plt.ylabel("Critic Loss")
plt.title("Critic Loss per Agent")
plt.legend()
plt.show()

# Win rates
plt.plot(blue_win_rate, label="Blue Team Win Rate")
plt.plot(orange_win_rate, label="Orange Team Win Rate")
plt.xlabel("Episodes")
plt.ylabel("Win Rate")
plt.title("Team Win Rates")
plt.legend()
plt.show()
```

```
import torch
import torch.nn as nn
import torch.optim as optim
from torch.distributions import Categorical
import numpy as np
import matplotlib.pyplot as plt
import soccer_twos

# ===============================
# 1. Actor–Critic Networks
# ===============================
class Actor(nn.Module):
    def __init__(self, obs_dim, action_branches):
        super().__init__()
        self.shared = nn.Sequential(
            nn.Linear(obs_dim, 128), nn.ReLU(),
            nn.Linear(128, 128), nn.ReLU()
        )
        self.heads = nn.ModuleList([nn.Linear(128, n) for n in action_branches])

    def forward(self, x):
        x = self.shared(x)
        return [torch.softmax(h(x), dim=-1) for h in self.heads]


class Critic(nn.Module):
    def __init__(self, obs_dim):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(obs_dim, 256), nn.ReLU(),
            nn.Linear(256, 256), nn.ReLU(),
            nn.Linear(256, 1)
        )

    def forward(self, x):
        return self.net(x)


# ===============================
# 2. IPPO Agent
# ===============================
class IPPO:
    def __init__(self, obs_dim, action_branches, num_agents, device="cpu",
                 gamma=0.99, lr=3e-4, batch_size=512, epochs=5,
                 clip=0.2, ent_coef=0.01, vf_coef=0.5):

        self.num_agents = num_agents
        self.device = torch.device(device)

        self.actors = [Actor(obs_dim, action_branches).to(self.device) for _ in range(num_agents)]
        self.critics = [Critic(obs_dim).to(self.device) for _ in range(num_agents)]
        self.actor_opts = [optim.Adam(a.parameters(), lr=lr) for a in self.actors]
        self.critic_opts = [optim.Adam(c.parameters(), lr=lr) for c in self.critics]

        self.gamma, self.batch_size, self.epochs = gamma, batch_size, epochs
        self.clip, self.ent_coef, self.vf_coef = clip, ent_coef, vf_coef

        self.memory = [[] for _ in range(num_agents)]
        self.actor_losses = [[] for _ in range(num_agents)]
        self.critic_losses = [[] for _ in range(num_agents)]

    def select_action(self, obs):
        actions, log_probs, values = [], [], []
        obs = torch.tensor(obs, dtype=torch.float32).to(self.device)
        for i in range(self.num_agents):
            dists = self.actors[i](obs[i])
            branch_actions, branch_log_probs = [], []
            for dist in dists:
                dist = Categorical(dist)
                a = dist.sample()
                branch_actions.append(a.item())
                branch_log_probs.append(dist.log_prob(a))
            actions.append(branch_actions)
            log_probs.append(torch.stack(branch_log_probs).sum())
            values.append(self.critics[i](obs[i]))
        return np.array(actions), log_probs, values

    def store(self, agent_idx, transition):
        self.memory[agent_idx].append(transition)

    def learn(self):
        total_actor_loss, total_critic_loss = 0.0, 0.0
        total_updates = 0

        for agent_idx in range(self.num_agents):
            mem = self.memory[agent_idx]
            if len(mem) < self.batch_size:
                continue

            obs = torch.tensor(np.array([m["obs"] for m in mem]), dtype=torch.float32).to(self.device)
            log_probs_old = torch.stack([m["log_prob"] for m in mem]).detach().to(self.device)
            values = torch.cat([m["value"] for m in mem]).squeeze().detach().to(self.device)
            rewards = torch.tensor(np.array([m["reward"] for m in mem]), dtype=torch.float32).to(self.device)
            dones = torch.tensor(np.array([m["done"] for m in mem]), dtype=torch.float32).to(self.device)

            # GAE
            advantages, gae = [], 0
            for t in reversed(range(len(rewards))):
                next_v = 0 if t == len(rewards)-1 else values[t+1]
                delta = rewards[t] + self.gamma * next_v * (1 - dones[t]) - values[t]
                gae = delta + self.gamma * 0.95 * (1 - dones[t]) * gae
                advantages.insert(0, gae)
            advantages = torch.tensor(advantages, dtype=torch.float32).to(self.device)
            returns = advantages + values

            # PPO updates
            for _ in range(self.epochs):
                idx = np.random.permutation(len(mem))
                for start in range(0, len(mem), self.batch_size):
                    batch = idx[start:start+self.batch_size]
                    batch_obs = obs[batch]
                    batch_adv = advantages[batch]
                    batch_ret = returns[batch]
                    batch_log_probs_old = log_probs_old[batch]

                    new_log_probs, entropies = [], []
                    for i in batch:
                        dists = self.actors[agent_idx](torch.tensor(mem[i]["obs"], dtype=torch.float32).to(self.device))
                        lp_per_branch = []
                        for k, dist_probs in enumerate(dists):
                            dist = Categorical(dist_probs)
                            a = torch.tensor(mem[i]["actions"][k]).to(self.device)
                            lp_per_branch.append(dist.log_prob(a))
                            entropies.append(dist.entropy().mean())
                        new_log_probs.append(torch.stack(lp_per_branch).sum())
                    new_log_probs = torch.stack(new_log_probs)

                    ratio = torch.exp(new_log_probs - batch_log_probs_old)
                    surr1 = ratio * batch_adv
                    surr2 = torch.clamp(ratio, 1 - self.clip, 1 + self.clip) * batch_adv
                    actor_loss = -torch.min(surr1, surr2).mean()

                    values_pred = self.critics[agent_idx](batch_obs).squeeze()
                    critic_loss = nn.MSELoss()(values_pred, batch_ret)

                    entropy_loss = torch.stack(entropies).mean()
                    loss = actor_loss + self.vf_coef * critic_loss - self.ent_coef * entropy_loss

                    self.actor_opts[agent_idx].zero_grad()
                    self.critic_opts[agent_idx].zero_grad()
                    loss.backward()
                    self.actor_opts[agent_idx].step()
                    self.critic_opts[agent_idx].step()

                    self.actor_losses[agent_idx].append(actor_loss.item())
                    self.critic_losses[agent_idx].append(critic_loss.item())

                    total_actor_loss += actor_loss.item()
                    total_critic_loss += critic_loss.item()
                    total_updates += 1

            self.memory[agent_idx] = []

        # Return mean losses across all agents for this epoch
        if total_updates > 0:
            return total_actor_loss / total_updates, total_critic_loss / total_updates
        else:
            return None, None


# ===============================
# 3. Training Loop
# ===============================
env = soccer_twos.make(render=False)
obs_dim, action_branches = env.observation_space.shape[0], env.action_space.nvec
num_agents = 4

device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"[INFO] Using device: {device}")
agent = IPPO(obs_dim, action_branches, num_agents, device=device)

episodes = 50
team_rewards = {"blue": [], "orange": []}
blue_cum_goals, orange_cum_goals = [], []
blue_goals_total, orange_goals_total = 0, 0
GOAL_THRESH = -1.5

for ep in range(episodes):
    obs = env.reset()
    done, ep_rew = {"__all__": False}, np.zeros(num_agents)

    while not done["__all__"]:
        obs_arr = np.stack([obs[i] for i in range(num_agents)])
        actions, log_probs, values = agent.select_action(obs_arr)
        action_dict = {i: np.array(actions[i]) for i in range(num_agents)}
        next_obs, reward, done, _ = env.step(action_dict)

        for i in range(num_agents):
            agent.store(i, {
                "obs": obs_arr[i],
                "actions": actions[i],
                "log_prob": log_probs[i],
                "value": values[i],
                "reward": reward[i],
                "done": float(done["__all__"]),
            })

        obs = next_obs
        ep_rew += np.array(list(reward.values()))

    blue_total, orange_total = ep_rew[0]+ep_rew[1], ep_rew[2]+ep_rew[3]
    team_rewards["blue"].append(blue_total)
    team_rewards["orange"].append(orange_total)

    blue_goals_ep, orange_goals_ep = 0, 0
    if blue_total <= GOAL_THRESH and orange_total >= 0: orange_goals_ep = 1
    elif orange_total <= GOAL_THRESH and blue_total >= 0: blue_goals_ep = 1
    blue_goals_total += blue_goals_ep
    orange_goals_total += orange_goals_ep
    blue_cum_goals.append(blue_goals_total)
    orange_cum_goals.append(orange_goals_total)

    actor_loss_mean, critic_loss_mean = agent.learn()

    # Print epoch summary including losses
    if actor_loss_mean is not None:
        total_loss = actor_loss_mean + critic_loss_mean
        print(f"[Ep {ep+1:03d}] Blue={blue_total:.3f}, Orange={orange_total:.3f} | "
              f"Actor={actor_loss_mean:.4f}, Critic={critic_loss_mean:.4f}, Total={total_loss:.4f} | "
              f"BlueGoals+={blue_goals_ep}, OrangeGoals+={orange_goals_ep}")
    else:
        print(f"[Ep {ep+1:03d}] (Not enough samples yet)")

env.close()

# ===============================
# 4. Loss & Team Trend Visualization
# ===============================
def smooth_curve(x, w=50):
    if len(x) < w: return np.array(x)
    return np.convolve(x, np.ones(w)/w, mode="valid")

# ---- (A) Actor Loss ----
plt.figure(figsize=(10,5))
for i in range(num_agents):
    plt.plot(smooth_curve(agent.actor_losses[i], 50), label=f"Actor Loss Agent {i}")
plt.xlabel("Update Steps"); plt.ylabel("Actor Loss")
plt.title("Actor Loss per Agent (Smoothed)"); plt.legend(); plt.show()

# ---- (B) Critic Loss ----
plt.figure(figsize=(10,5))
for i in range(num_agents):
    plt.plot(smooth_curve(agent.critic_losses[i], 50), label=f"Critic Loss Agent {i}")
plt.xlabel("Update Steps"); plt.ylabel("Critic Loss")
plt.title("Critic Loss per Agent (Smoothed)"); plt.legend(); plt.show()

# ---- (C) Total Loss ----
plt.figure(figsize=(10,5))
for i in range(num_agents):
    total_loss = np.array(agent.actor_losses[i]) + np.array(agent.critic_losses[i])
    plt.plot(smooth_curve(total_loss, 50), label=f"Total Loss Agent {i}")
plt.xlabel("Update Steps"); plt.ylabel("Total Loss (Actor + Critic)")
plt.title("Combined Total Loss per Agent (Smoothed)")
plt.legend(); plt.show()

# ---- (D) Team-Level Mean Loss ----
plt.figure(figsize=(10,5))
blue_mean = (np.array(agent.actor_losses[0]) + np.array(agent.actor_losses[1])) / 2
orange_mean = (np.array(agent.actor_losses[2]) + np.array(agent.actor_losses[3])) / 2
plt.plot(smooth_curve(blue_mean, 50), label="Blue Team Avg Actor Loss")
plt.plot(smooth_curve(orange_mean, 50), label="Orange Team Avg Actor Loss")
plt.xlabel("Update Steps"); plt.ylabel("Mean Actor Loss")
plt.title("Team-Level Actor Loss Trend (Coordination Stability)")
plt.legend(); plt.show()

print("\n=== Training Diagnostics Complete ===")
```
